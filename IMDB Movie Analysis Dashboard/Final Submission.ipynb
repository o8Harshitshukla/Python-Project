**Assignment 14. Add a column ‘Votes_norm’ in the IMDB dataset using apply() function where Votes_norm is** <br>

**[Votes - min(Votes)]*10/[max(votes) - min(votes)]** <br>

**The above formula is the normalization formula and converts votes into a scale of 0-10.** <br><br>

**What is the average ‘Votes_norm’ ? (Round to two decimal places)** <br>

**NOTE: Use original dataset. (Before cleaning)**

```python
df["Votes_norm"] = df.apply(lambda x: (x["Votes"] - min(df["Votes"]))*10/(max(df["Votes"]) - min(df["Votes"])), axis=1)
df
```

```python
round(df["Votes_norm"].mean(),2)
```

**Create a new column ‘Revenue_bins’ so that ‘Revenue_millions’ are categorized in buckets 0-50, 51-100, 101-150, and 150+. Which bucket has the highest number of movies?**

```python
# Define the bucket ranges
buckets = {'0-50': (0, 50),
           '51-100': (51, 100),
           '101-150': (101, 150),
           '150+': (151, float('inf'))}

# Function to categorize revenue into buckets
def categorize_revenue(revenue):
    for bucket, (start, end) in buckets.items():
        if start <= revenue <= end:
            return bucket

# Categorize the 'Revenue_millions' column into buckets
df["Revenue_bins"] = df["Revenue_millions"].apply(categorize_revenue)

df
```

**OR ( With Cut Function )**

```python
# Define the bucket ranges
bins = [0, 50, 100, 150, float('inf')]
labels = ['0-50', '51-100', '101-150', '150+']

# Categorize the 'Revenue_millions' column into buckets
df["Revenue_bins"] = pd.cut(df["Revenue_millions"], bins=bins, labels=labels)

df
```

```python
df.groupby(by=["Revenue_bins"])["ID"].count()
```
